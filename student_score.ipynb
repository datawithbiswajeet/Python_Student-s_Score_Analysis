```python

import pandas as pd

df = pd.read_csv("student_scores.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df = df.drop(columns = "Unnamed: 0")
df
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].replace("05-Oct", "5-10")
df
```

```python
df.dropna()         # removes the rows which contains null value
```

```python
df["EthnicGroup"].fillna("unknown")
```

```python
df["NrSiblings"].mean()
df["NrSiblings"].median()
df["NrSiblings"].mode()
```

```python
df["NrSiblings"] = df["NrSiblings"].fillna(df["NrSiblings"].median())
df["NrSiblings"].bfill()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(5,3))
sns.histplot(df['MathScore'], kde=True, bins=10, color='skyblue')
plt.title("Distribution of Student Math Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['ReadingScore'], kde=True, bins=15, color='red')
plt.title("Distribution of Student Reading Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['WritingScore'], kde=True, bins=20, color='orange')
plt.title("Distribution of Student Writing Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.boxplot(x='Gender', y='WritingScore', data=df)
plt.title("Math Score by Gender")
plt.show()
```

**This Shows Girl ðŸ‘© are always good at writing skills ðŸ˜€**

```python
sns.countplot(x='EthnicGroup', data=df)
plt.title("Distribution by Ethnic Group")
plt.show()
```

```python
sns.barplot(x='ParentEduc', y='ReadingScore', data=df)
plt.xticks(rotation=45)
plt.title("Reading Score vs Parent Education")
plt.show()
```

```python
df[['MathScore', 'ReadingScore', 'WritingScore']].hist(bins=10, figsize=(12,5), color='orange' )
plt.suptitle("Score Distributions")
plt.show()
```

```python
score_ranges = pd.cut(df['MathScore'], bins=[0, 40, 60, 80, 100], labels=['Poor','Average','Good','Excellent'])
score_ranges.value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title('Math Score Range Distribution')
plt.show()
```

```python
sns.boxplot(x='WklyStudyHours', y='MathScore', data=df)
plt.title("Math Score by Study Hours")
plt.show()
```
